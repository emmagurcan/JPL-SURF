```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr                     
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import glob
from datetime import date, time, datetime 
import pandas as pd
import netCDF4 as net
import os
from os import listdir
from os.path import join, isfile
```

```python
def plot_swot_output(filename):
    model_dset = xr.open_dataset(filename)
    model_ssh = model_dset['SSH_obs']
    model_ssh = np.where(model_ssh<0, np.NaN, model_ssh)
    
    lon = model_dset['lon']
    lat = model_dset['lat']
    
    plt.scatter(lon, lat, c=model_ssh, s=2, cmap='coolwarm')
```

```python
def plot_all_files(data_files, pathname, num):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
    gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                   linewidth=.5, color='gray', alpha=0.5, linestyle='--')
    gl.top_labels = False
    gl.right_labels = False 

    for f in data_files:
        name = os.path.join(pathname, f)
        plot_swot_output(name)

    # coastlines
    ax.add_feature(cfeature.LAND)
    ax.add_feature(cfeature.COASTLINE)

    # colorbar
    plt.colorbar(shrink = 0.7)

    # set title
    title = 'Synthetic SWOT C0{n}Nadir Absolute Dynamic Topography (in meters)'
    plt.title(title.format(n=num))

    plt.show()
```

```python
# C01
foldername = '/Users/emmagurcan/Documents/JPL_2021/Swot_Simulator_Code/model_nadir_swot_ouput/Nadir_C01'
C01data_files = [f for f in listdir(foldername) if isfile(join(foldername, f))]
plot_all_files(C01data_files, foldername, 1)
```

```python
# C02
foldername_2 = '/Users/emmagurcan/Documents/JPL_2021/Swot_Simulator_Code/model_nadir_swot_ouput/Nadir_C02'
C02data_files = [f for f in listdir(foldername_2) if isfile(join(foldername_2, f))]
plot_all_files(C02data_files, foldername_2, 2)
```

```python
# C03
foldername_3 = '/Users/emmagurcan/Documents/JPL_2021/Swot_Simulator_Code/model_nadir_swot_ouput/Nadir_C03'
C03data_files = [f for f in listdir(foldername_3) if isfile(join(foldername_3, f))]
plot_all_files(C03data_files, foldername_3, 3)
```
